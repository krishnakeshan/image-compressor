```
import os

import numpy as np

import tensorflow as tf
from tensorflow import keras

from keras import models, layers
```

```
# extract image files
import zipfile

with zipfile.ZipFile('coil-100.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```
# read and store images
from PIL import Image

image_width, image_height, channels = 128, 128, 3

images_directory = '/content/coil-100'
filenames = os.listdir(images_directory)

train_x = []

# retrieve pixel data of each image
for filename in filenames[:5000]:
    if not filename.endswith('.png'):
        continue
    im = Image.open(os.path.join(images_directory, filename))
    train_x.append(list(im.getdata()))

train_x = np.array(train_x)
train_y = train_x.reshape(5000, image_width * image_height * channels)
```

```
# create network
model = models.Sequential()
model.add(layers.Flatten(input_shape=(image_width * image_height, channels)))
model.add(layers.Dense(300, activation='relu'))
model.add(layers.Dense(300, activation='relu'))
model.add(layers.Dense(image_width * image_height * channels, activation='relu'))

model.compile('rmsprop', 'mse', metrics=['accuracy'])
```

```
model.fit(train_x, train_y, epochs=70, batch_size=20)
```
